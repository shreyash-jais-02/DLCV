# Canny Edge Detection

```python
#importing the required Packages for Canny Edge Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image
image_path = r'D:\Educational\Fall Semester 23-24\Deep Learning for Computer Vision\Lab Codes\Lion.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Check if the image was loaded successfully
if image is None:
    print("Error: Image not found or could not be loaded.")
else:
    # Perform Canny edge detection
    edges = cv2.Canny(image, 100, 200)  # You can adjust the threshold values as needed

    # Display the original image and the detected edges using matplotlib
    plt.figure(figsize=(12, 6))

    plt.subplot(121)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(122)
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edge Detection')
    plt.axis('off')

    plt.show()
```
